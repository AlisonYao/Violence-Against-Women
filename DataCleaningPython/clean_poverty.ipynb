```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('poverty.csv')
df.head()
```

```python
df.info()
```

```python
for index, row in df.iterrows():
    poverty_headcount_ratio = np.nan
    for year in range(2020, 1984, -1):
        # print(index, row['CountryName'], year, row[str(year)], )
        if not pd.isnull(row[str(year)]):
            poverty_headcount_ratio = row[str(year)]
            break
    # print(index, row['CountryName'], poverty_headcount_ratio)
    df.at[index, 'poverty_headcount_ratio'] = poverty_headcount_ratio
    # print(index, row['CountryName'], row['1984'])
```

```python
df.head()
```

```python
output = df[['CountryName', 'Country', 'poverty_headcount_ratio']]
```

```python
output.to_csv(',,/Data/poverty_cleaned.csv', index=False)
```
